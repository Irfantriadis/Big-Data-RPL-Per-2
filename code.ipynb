```python
import re
import nltk
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def preprocess_text(text):
    # 1. Mengubah teks menjadi huruf kecil
    text = text.lower()
    
    # 2. Menghilangkan URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # 3. Menghilangkan username (mention) dan hashtag
    text = re.sub(r'@\w+|#\w+', '', text)
    
    # 4. Menghilangkan angka
    text = re.sub(r'\d+', '', text)
    
    # 5. Menghilangkan karakter spesial dan tanda baca
    text = re.sub(r'[^\w\s]', '', text)
    
    # 6. Menghilangkan stopwords bahasa Indonesia
    text = ' '.join([word for word in text.split() if word not in stop_words])
    
    # 7. Stemming (mengembalikan kata ke bentuk dasar)
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    
    return text
```

```python
df['Tweet'] = df['Text Tweet'].apply(preprocess_text)
```

```python
print("Tweet Asli:")
print(df['Text Tweet'].head())
print("\nTweet Setelah Preprocessing:")
print(df['Tweet'].head())
```

```python
singkatan_dict = {
    'yg': 'yang',
    'gk': 'tidak',
    'ga': 'tidak',
    'tdk': 'tidak',
    'dr': 'dari',
    'dgn': 'dengan',
    'kalo': 'kalau',
    'tp': 'tapi',
    'utk': 'untuk',
    'pd': 'pada',
    'dg': 'dengan',
    'rmh': 'rumah',
    'dmn': 'dimana',
    'sblm': 'sebelum',
    'sy': 'saya',
    'dlm': 'dalam',
    'kl': 'kalau',
    'sdh': 'sudah',
    'trs': 'terus',
    'dpt': 'dapat',
    'jkt': 'jakarta'
    # Tambahkan singkatan lain sesuai kebutuhan
}
```

```python
def replace_singkatan(text):
    return ' '.join([singkatan_dict[word] if word in singkatan_dict else word for word in text.split()])
```

```python
def preprocess_singkatan(text):
    text = replace_singkatan(text)    
    return text
```

```python
df['Tweet'] = df['Tweet'].apply(preprocess_singkatan)
```
